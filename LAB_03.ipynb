# 4. Create different chart types: (4)
- Line Plot
- Bar Plot
- Histogram

```python
# Line plot
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]

plt.figure(figsize=(10, 5))
plt.plot(x, y, color='red', marker='o')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plot')
plt.show()

# Bar plot
bar_heights = [5, 3, 8, 4, 6]

plt.figure(figsize=(10, 5))
plt.bar(x, bar_heights, color='green')
plt.xlabel('x')
plt.ylabel('Bar heights')
plt.title('Bar Plot')
plt.show()

# Histogram
data = []
for i in range(len(x)):
    data.extend([x[i]] * y[i])

plt.figure(figsize=(10, 5))
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

# 5. Enhance plot with annotations: (4)
- Add an annotation on a random point in plot

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]

plt.figure(figsize=(10, 5))
plt.plot(x, y, color='red', marker='o')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plot')

index = np.random.randint(0, len(x))

plt.annotate(f'Point {index+1}', (x[index], y[index]), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()

```

# 6. Subplots and multiple plots: (4)
Create 2 Subplots
+ Bar Plot
+ Scatter Plot

```python

x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]
bar_heights = [5, 3, 8, 4, 6]

plt.figure(figsize=(10, 10))

# Scatter plot
plt.subplot(2, 1, 1)
plt.scatter(x, y, color='blue', marker='o')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')

# Bar plot
plt.subplot(2, 1, 2)
plt.bar(x, bar_heights, color='green')
plt.xlabel('x')
plt.ylabel('Bar heights')
plt.title('Bar Plot')

plt.tight_layout()
plt.show()
```

# 9. Layout and composition: (4)

Explore the impact of different plot sizes and aspect ratios by creating a small square plot and a larger wide plot using the 'scatter' function.
- Small = 5x5
- Large = 10x5

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]
colors = ['black', 'blue', 'purple', 'black', 'blue']
sizes = 100 * np.random.rand(5)

# Create a small square plot (5x5)
plt.figure(figsize=(5, 5))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.title('Small Square Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Create a larger wide plot (10x5)
plt.figure(figsize=(10, 5))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.title('Large Wide Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Modify the legend placement to the upper left and adjust the positioning of the annotation for the point (2, 5) in a scatter plot

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]

plt.scatter(x, y, color='blue', label='Data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.legend(loc='upper left')
# Adjust legend placement
plt.annotate('Point (2, 5)', (2, 5), textcoords="offset points", xytext=(5,-5), ha='center')
# Adjust annotation position
plt.show()
```

Enhance the use of whitespace in a plot by adjusting the spacing around the plot area and creating a scatter plot with improved whitespace.

```python
x = [1, 2, 3, 4, 5]
bar_heights = [5, 3, 8, 4, 6]

plt.figure(figsize=(10, 5))
plt.bar(x, bar_heights, color='green')
plt.xlabel('x')
plt.ylabel('Bar heights')
plt.title('Bar Plot')

# Adjust spacing and margins
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

plt.show()
```

Choose appropriate chart type for the following data

# 11. Customize plot styles: (4)

Experiment with line styles, marker shapes, and color gradients:
- Dashed Line
- Triangle Marker (Green)

Utilize color gradients to represent data using a specified colormap, for example, 'YlGnBu', to create a scatter plot with a color bar.

```python
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 2, 1]
colors = range(len(x))  # Use index for color gradient
sizes = 1000 * np.random.rand(len(x))

plt.figure(figsize=(8, 6))

# Create scatter plot with color gradient and color bar
scatter = plt.scatter(x, y, c=colors, s=sizes, cmap='YlGnBu', marker='^')
plt.colorbar(scatter, label='Index')

# Create dashed line
plt.plot(x, y, linestyle='--', color='black')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Line, Triangle Marker, and Color Gradient')
plt.show()

```
